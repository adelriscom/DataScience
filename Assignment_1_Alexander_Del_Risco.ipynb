```python
def check_variables(df: pd.DataFrame):
  cols = df.columns
  quantitative_cols = df._get_numeric_data().columns
  categorical_cols = list(set(cols) - set(quantitative_cols))
  return categorical_cols, list(set(quantitative_cols))

cat_cols, quan_cols = check_variables(data_f)
print(f'categorical columns: {cat_cols}')
print(f'quantitative columns: {quan_cols}')
```

>>Q8) Examine distribution of age variable 
- Plot histogram for age variable
- Interpret your observation from the histogram with respect to the shape,
center, spread, and outliers.
- Calculate measures of center of the distribution of age: mode, mean, median
- Calculate measures of spread of the distribution of age: range, IQR, outliers
based on 1.5IQR criterion, standard deviation.
- Plot the box plot for age variable
- Provide your summary/interpretation about the distribution of age.

```python
#histogram for the age variable
plt.hist(data_f['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

>>The histogram is right-skewed, with the majority of individuals falling in the age range of around 20 to 50 years old. There are some individuals over the age of 70, but they are relatively few in number compared to the rest of the distribution.

```python
# Calculate measures of center of the distribution of age
data_f['age'].mode()
data_f['age'].mean()
data_f['age'].median()
```

```python
min_ = data_f['age'].min()
max_ = data_f['age'].max()
range_ = max_ - min_
print(range_)
```

```python
Q25 = data_f['age'].quantile(0.25)
Q75 = data_f['age'].quantile(0.75)
IQR = Q75 - Q25

age_outliers = data_f[(data_f.age < Q25 - 1.5*IQR) |
(data_f.age > Q75 + 1.5*IQR)]['age']

age_std = data_f['age'].std()

plt.boxplot(x=data_f['age'], labels=['age'])
plt.ylabel('Age')
```

>>Based on the box plot, we can see that the median age is around 37, with the IQR spanning from around 25 to 50 years old. There are also several outliers above the upper whisker, indicating that there are some individuals in the dataset who are significantly older than the rest of the population.
