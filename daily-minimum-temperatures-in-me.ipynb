```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Read your data into a DataFrame (replace 'your_data.csv' with your file path)
df = pd.read_csv('/Users/alexanderdelriscomorales/Downloads/data_tsa/Electric_Production.csv')

# Calculate autocorrelation values
lags = 40  # Number of lags to calculate
acf_values = sm.tsa.acf(df['IPG2211A2N'], nlags=lags)

# Create a lag array for the x-axis
lags_array = range(lags + 1)

# Plot the autocorrelation function
plt.figure(figsize=(10, 6))
plt.bar(lags_array, acf_values)
plt.title('Autocorrelation Function')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()

```
